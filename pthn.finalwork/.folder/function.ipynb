Даны две функции: f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
Требуется:

Найти координаты точек пересечения
Построить графики функций в одной системе координат
Построить график функции пересечения

```python
from sympy import *
x = Symbol('x')
ax = x**3 - 50*x
bx = -1*x**4 + 88*x**2 - 241
abx = x**3 - 50*x + 1*x**4 - 88*x**2 + 241
```

Вычисление корней выражений

```python
roots = solve(Eq(ax, bx), x)
print(roots)
```

Словарь с х и у

```python
cross_dict = {}
c = 1
for z in roots:
  cross_dict[c] = {'x': round(N(z), 5), 'y': round(N(-1*z**4 + 88*z**2 - 241), 5)}
  c += 1
```

```python
cross_dict
```

Вывод координат точек пересечения

```python
print(*[(cross_dict[i]['x'], cross_dict[i]['y']) for i in cross_dict.keys()], sep=', ')
```

Вывод графиков функций в системе координат

```python
from sympy.plotting import plot
plot(ax, bx, x)
```

График функций пересечения

```python
plot(abx, x)
```

```python
p = plot(ax, bx, abx, show=false)
p[2].line_color = 'r'
p.show()
```
